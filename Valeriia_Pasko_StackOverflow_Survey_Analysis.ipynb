```python
schema = pd.read_csv('/Users/lerapasko/Downloads/schema.csv')
schema.head()
```

```python
# Отримуємо список питань з колонки qname
schema_questions = set(schema["qname"])

# Отримуємо список колонок основного датасету
df_columns = set(df.columns)

# Інтерсекція множин
existing_questions = schema_questions.intersection(df_columns)

# Фільтруємо тільки потрібні колонки
filtered_df = df[list(existing_questions)]

# Знаходимо респондентів без пропущених значень
complete_respondents = filtered_df.notna().all(axis=1)

# Підрахунок кількості таких респондентів
total_complete = complete_respondents.sum()

print("Кількість респондентів, які відповіли на всі питання:")
print(total_complete)
```

```python
# Вибираємо колонку WorkExp
work_exp = df["WorkExp"]

# Обчислення мір центральної тенденціїб, даний спосіб ігнорує NaN значення 
mean_value = work_exp.mean()
median_value = work_exp.median()
mode_value = work_exp.mode()[0]

# Формуємо результати у вигляді Series
central_tendency = pd.Series({
    "Mean": round(mean_value, 3),
    "Median": round(median_value, 3),
    "Mode": round(mode_value, 3)
})

print("Міри центральної тенденції для WorkExp:")
print(central_tendency)
```

```python
# Фільтруємо респондентів, які програмують на Python
python_devs = df[df["LanguageHaveWorkedWith"].str.contains( "Python", case=False, na=False)]

# Видаляємо рядки з пропущеною компенсацією або країною
python_devs = python_devs.dropna(
    subset=["Country", "ConvertedCompYearly"]
)

# Групування по країнах та обчислення стати
compensation_stats = (
    python_devs
    .groupby("Country")["ConvertedCompYearly"]
    .agg(
        MeanCompensation="mean",
        MedianCompensation="median"
    )
    .reset_index()
)

display(compensation_stats)
```

```python
# Видаляємо рядки з пропущеними значеннями
top_paid = df.dropna(subset=["ConvertedCompYearly", "EdLevel"])

# Сортуємо за компенсацією у спадному порядку
top_paid = top_paid.sort_values(by="ConvertedCompYearly",ascending=False)

# Обираємо топ-5 респондентів
top_5 = top_paid.head(5)

result = top_5[["ConvertedCompYearly", "EdLevel"]]

display(result)
```

```python
# Створюємо колонку з інформацією,чи використовує респондент Python
df["UsesPython"] = df["LanguageHaveWorkedWith"].str.contains("Python", case=False,na=False)

# Видаляємо рядки без вікової категорії
age_data = df.dropna(subset=["Age"])

# Групуємо за віковими категоріями
python_by_age = (
    age_data
    .groupby("Age")
    .agg(
        PythonPercent=("UsesPython", lambda x: x.mean() * 100)
    )
    .reset_index()
)

# Перейменовуємо колонки
python_by_age.columns = ["AgeCategory","PythonDevelopersPercent"]

# Округлюємо значення
python_by_age["PythonDevelopersPercent"] = (
    python_by_age["PythonDevelopersPercent"]
    .round(2)
)

# Виводимо результат
display(python_by_age)
```
